```python
import pandas as pd
import json

with open('results_plex.json') as f:
    df=json.load(f)['PLEX']
names={'up':'1-up','down':'2-down','weighted':'3-combined'}
data=dict()
for i in ['2','46']:
    data[i]=[]
    for res in df[i]:
        # line={'r':i}
        line={}
        losses=res['losses']
        f_loss=dict()
        way=res['way']
        if way=='both' :
            for metric in ['up','down','weighted']:
                line={}
                losses=res['losses']
                for k in list(losses.keys()):
                    if not 'up' in k:
                        losses.pop(k)
                f_loss=dict()
                line['dice']=res['dice_'+metric]*100
                line['hauss']=res['hauss_'+metric]
                line['asd']=res['asd_'+metric]
                # line['bidir-cons-reg']=losses['bidir-cons-reg']
                # line['bidir-cons-dice']=losses['bidir-cons-dice']
                line['way']=names[metric]
                # del res['losses']
                line.update(losses)
                # for i,way in enumerate(['up','down','combined']):
                #     if way==line['way']: line['way']=f'{i}-{way}'
                data[i].append(line)
print(data['2'])
# for i in df.keys():
truc=pd.DataFrame(data['2'])
truc.sort_values('way',ascending=False,inplace=True)

table=truc.groupby(['way'],sort=True).mean()
table
for i in ['46']:	
    if i!='2':
        truc=pd.DataFrame(data[i])
        truc.sort_values('way',ascending=False,inplace=True)

        truc=truc.groupby(['way','compo-reg-up','compo-dice-up'],sort=True).mean()
        table=truc#pd.concat([table,truc],axis=1)
table.to_excel('res_plex_1on20.xlsx')
```
