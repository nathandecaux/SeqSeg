```python
import pandas as pd
import json

with open('results.json') as f:
    df=json.load(f)['PLEX']
names={'up':'1-up','down':'2-down','combined':'3-combined'}
data=dict()
for i in df.keys():
    data[i]=[]
    for res in df[i]:
        # line={'r':i}
        line={}
        losses=res['losses']
        f_loss=dict()
        for k in ['compo-reg-','compo-dice-']:
            way=res['way']
            if way=='combined' or way=='both':
                metrics='weighted' 
                way='up'
            else:
                metrics=way
            line['dice']=res['dice_'+metrics]
            line['hauss']=res['hauss_'+metrics]
            line['asd']=res['asd_'+metrics]
            # line['bidir-cons-reg']=losses['bidir-cons-reg']
            # line['bidir-cons-dice']=losses['bidir-cons-dice']
            f_loss[k[:-1]]=losses[k+way]

        if res['way']!='both':
            
            line['way']=names[res['way']]

            del res['losses']
            line.update(f_loss)
            # for i,way in enumerate(['up','down','combined']):
            #     if way==line['way']: line['way']=f'{i}-{way}'
            data[i].append(line)

# for i in df.keys():
truc=pd.DataFrame(data['2'])
truc.sort_values('way',ascending=False,inplace=True)

table=truc.groupby(['compo-reg','compo-dice','way'],sort=True).mean()

for i in df.keys():
    if i!='2':
        truc=pd.DataFrame(data['2'])
        truc.sort_values('way',ascending=False,inplace=True)

        truc=truc.groupby(['compo-reg','compo-dice','way'],sort=True).mean()
    # truc.to_excel(f'res_{i}.xlsx')
```
